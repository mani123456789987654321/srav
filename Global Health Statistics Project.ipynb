This program creates a **4-panel bar chart visualization** using data from a DataFrame (`df`). Each panel focuses on a different categorical variable, providing insights into their distribution. Here's what it does in detail:

1. **DataFrame Structure**:
   - The dataset contains columns for `Gender`, `Treatment Type`, `Country`, `Hospital Beds per 1000`, and `Disease Category`.

2. **Visualizations**:
   - The program uses **subplots** to create four bar charts in a 2x2 grid layout.
   - Each bar chart represents the count of unique values for a specific categorical variable.

3. **Details of Each Plot**:
   - **Plot 1: Gender**  
     Visualizes the distribution of individuals by gender (`Male`, `Female`).
   - **Plot 2: Treatment Type**  
     Shows the count of individuals for each treatment type (e.g., `Type1`, `Type2`, `Type3`).
   - **Plot 3: Country**  
     Displays the count of individuals grouped by their respective countries (`USA`, `India`, `Canada`).
   - **Plot 4: Disease Category**  
     Highlights the distribution of individuals across disease categories (`Infectious`, `Non-Communicable`).

4. **Customization**:
   - **Distinct Colors**: Each plot uses unique colors for visual differentiation.
   - **Titles**: Added descriptive titles for each subplot.
   - **Formatting**:
     - X-axis labels are rotated by 90 degrees for readability.
     - Y-axis limits are adjusted dynamically to accommodate all bars.

5. **Layout Adjustment**:
   - **`plt.tight_layout()`** ensures proper spacing between subplots.
   - The `rect=[0, 0.07, 1, 1]` parameter provides additional space for subplot titles and ensures no overlap.

### **Purpose**:
This program provides a quick and visually appealing summary of key categorical variables in the dataset, allowing users to compare distributions and spot patterns or anomalies.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example dataframe structure
df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'Treatment Type': ['Type1', 'Type2', 'Type1', 'Type3'],
    'Country': ['USA', 'India', 'USA', 'Canada'],
    'Hospital Beds per 1000': [3, 2, 1, 3],
    'Disease Category': ['Infectious', 'Non-Communicable', 'Infectious', 'Non-Communicable']
})

# Define distinct colors for each bar plot
colors1 = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
colors2 = ['#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
colors3 = ['#17becf', '#bcbd22', '#7f7f7f', '#e377c2', '#8c564b']
colors4 = ['#d62728', '#2ca02c', '#ff7f0e', '#1f77b4', '#9467bd']

plt.figure(figsize=(16, 12))

# Plot 1: Gender
plt.subplot(221)
gender_counts = df['Gender'].value_counts()
gender_counts.plot(kind='bar', title='Gender of People', color=colors1[:gender_counts.nunique()])
plt.xticks(rotation=90)
plt.ylim(0, gender_counts.max() + 1)
plt.title('Gender of People', fontsize=14, pad=20)

# Plot 2: Treatment Type
plt.subplot(222)
treatment_counts = df['Treatment Type'].value_counts()
treatment_counts.plot(kind='bar', title='Treatment Type of Disease for People', color=colors2[:treatment_counts.nunique()])
plt.xticks(rotation=90)
plt.ylim(0, treatment_counts.max() + 1)
plt.title('Treatment Type of Disease for People', fontsize=14, pad=20)

# Plot 3: Country
plt.subplot(223)
country_counts = df['Country'].value_counts()
country_counts.plot(kind='bar', title='Country of Individuals', color=colors3[:country_counts.nunique()])
plt.xticks(rotation=90)
plt.ylim(0, country_counts.max() + 1)
plt.title('Country of Individuals', fontsize=14, pad=20)

# Plot 4: Disease Category
plt.subplot(224)
disease_category_counts = df['Disease Category'].value_counts()
disease_category_counts.plot(kind='bar', title='Disease Category', color=colors4[:disease_category_counts.nunique()])
plt.xticks(rotation=90)
plt.ylim(0, disease_category_counts.max() + 1)
plt.title('Disease Category', fontsize=14, pad=20)
# Move the text higher for better placement


plt.tight_layout(rect=[0, 0.07, 1, 1])  # Adjust layout to create space for descriptions
plt.show()
```

This program visualizes various aspects of healthcare data using multiple plots:  

1. **Doctors per 1000**:  
   - A histogram shows the frequency distribution of doctors per 1000 people (range: 0-5).  
   - A KDE plot provides a smooth density estimate for the same variable.  

2. **Doctors by Gender**:  
   - A boxplot compares the distribution of doctors per 1000 people across genders (`Male`, `Female`).  

3. **Correlation Analysis**:  
   - A heatmap displays the correlation between all numeric variables in the dataset, revealing relationships among key factors like prevalence, incidence, and recovery rates.  

These visualizations help explore data distributions, gender-based variations, and inter-variable relationships.

```python


import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data with additional numeric columns
data = {
    'Doctors per 1000': np.random.uniform(0, 5, 100),
    'Gender': np.random.choice([0, 1], size=100),
    'Average Treatment Cost (USD)': np.random.uniform(1000, 5000, 100),
    'Improvement in 5 Years (%)': np.random.uniform(0, 20, 100),
    'Prevalence Rate (%)': np.random.uniform(5, 25, 100),
    'Incidence Rate (%)': np.random.uniform(1, 15, 100),
    'Mortality Rate (%)': np.random.uniform(0, 10, 100),
    'Hospital Beds per 1000': np.random.uniform(1, 5, 100),
    'Recovery Rate (%)': np.random.uniform(50, 90, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Ensure Gender is properly categorized
df['Gender'] = df['Gender'].map({0: 'Male', 1: 'Female'}).fillna('Unknown')

# Distribution of Doctors per 1000 with different ranges
plt.figure(figsize=(10, 6))
sns.histplot(df['Doctors per 1000'], kde=True)
plt.title('Distribution of Doctors per 1000 (Range: 0-5)')
plt.xlabel('Doctors per 1000')
plt.ylabel('Frequency')
plt.xlim(0, 5)  # Set range for x-axis (0 to 5)
plt.tight_layout()
plt.show()

# Box plot of Doctors per 1000 by Gender with different range
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Doctors per 1000', data=df)
plt.title('Doctors per 1000 by Gender (Range: 0-5)')
plt.xlabel('Gender')
plt.ylabel('Doctors per 1000')
plt.ylim(0, 5)  # Set range for y-axis (0 to 5)
plt.tight_layout()
plt.show()

# Distribution plot (kde plot) for Doctors per 1000 with the updated `fill` parameter
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Doctors per 1000'], fill=True, color='orange')  # Use `fill=True` instead of `shade=True`
plt.title('Distribution Plot of Doctors per 1000')
plt.xlabel('Doctors per 1000')
plt.ylabel('Density')
plt.xlim(0, 5)  # Set range for x-axis
plt.tight_layout()
plt.show()

# Correlation matrix for all numeric columns
numeric_columns = df.select_dtypes(include=[np.number])  # Select only numeric columns
correlation_matrix = numeric_columns.corr()

# Heatmap with more values (correlations for many columns)
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of All Numeric Columns')
plt.tight_layout()
plt.show()
```

## Step 7: Outlier Detection
Identify and handle outliers using visualizations and statistical methods.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Box plot for outlier detection in Healthcare Access (%)
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Healthcare Access (%)'], color='#d62728')  # Custom color for the boxplot
plt.title('Box Plot for Healthcare Access (%)')
plt.show()

# Handling outliers (example: using IQR method)
q1 = df['Healthcare Access (%)'].quantile(0.25)
q3 = df['Healthcare Access (%)'].quantile(0.75)
iqr = q3 - q1

# Define outliers
outliers = df[(df['Healthcare Access (%)'] < (q1 - 1.5 * iqr)) | (df['Healthcare Access (%)'] > (q3 + 1.5 * iqr))]
print(outliers)
```

## Step 8: Hypothesis Testing
Formulate hypotheses based on initial observations and test them using statistical methods.

```python
from scipy.stats import ttest_ind

# Hypothesis: Male and female groups have different Healthcare Access (%)
male_healthcare_access = df[df['Gender'] == 0]['Healthcare Access (%)']
female_healthcare_access = df[df['Gender'] == 1]['Healthcare Access (%)']

# Perform independent t-test
t_stat, p_value = ttest_ind(male_healthcare_access, female_healthcare_access, nan_policy='omit')
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")


```
